```python
import numpy as np
import math
import re

class SimpleTextMatrixConverter:
    def __init__(self, width=30, padding_value=0, to_lowercase=True, normalize_whitespace=True):
        """
        Inicializa o conversor com configurações básicas.
        - width: Largura da matriz numérica.
        - padding_value: Valor usado para preencher a última linha da matriz, se necessário.
        - to_lowercase: Se True, converte o texto para minúsculas.
        - normalize_whitespace: Se True, remove pontuações (exceto letras e números),
                                substitui múltiplos espaços por um único e remove espaços nas bordas.
        """
        self.width = width
        self.padding_value = padding_value
        self.to_lowercase = to_lowercase
        self.normalize_whitespace = normalize_whitespace

    def _clean_text(self, text: str) -> str:
        """Limpa o texto de acordo com as configurações."""
        processed_text = text
        if self.to_lowercase:
            processed_text = processed_text.lower()

        if self.normalize_whitespace:
            # Mantém apenas letras, números e espaços. Remove o resto.
            processed_text = re.sub(r'[^a-z0-9\s]', '', processed_text) if self.to_lowercase else re.sub(r'[^a-zA-Z0-9\s]', '', processed_text)
            # Substitui múltiplos espaços por um único espaço
            processed_text = re.sub(r'\s+', ' ', processed_text)
            processed_text = processed_text.strip() # Remove espaços no início/fim
        return processed_text

    def text_to_matrix(self, text: str) -> tuple[np.ndarray, str]:
        """
        Converte o texto fornecido em uma matriz de seus valores ASCII.
        Retorna a matriz e o texto limpo utilizado.
        """
        cleaned_text = self._clean_text(text)
        if not cleaned_text: # Se o texto limpo estiver vazio
            return np.array([]).reshape(0, self.width), cleaned_text

        ord_values = [ord(c) for c in cleaned_text]

        num_elements = len(ord_values)
        num_rows = math.ceil(num_elements / self.width)
        total_matrix_elements = num_rows * self.width
        padding_count = total_matrix_elements - num_elements

        ord_values_padded = ord_values + [self.padding_value] * padding_count
        matrix = np.array(ord_values_padded, dtype=int).reshape(num_rows, self.width)

        return matrix, cleaned_text

    def matrix_to_text(self, matrix: np.ndarray) -> str:
        """
        Converte uma matriz de valores ASCII de volta para o texto (limpo).
        """
        if matrix.size == 0:
            return ""

        flat_values = matrix.flatten()

        # Encontra o último valor real (não padding)
        last_real_char_index = -1
        for i in range(len(flat_values) - 1, -1, -1):
            if flat_values[i] != self.padding_value:
                last_real_char_index = i
                break

        if last_real_char_index != -1:
            values_no_padding = flat_values[:last_real_char_index + 1]
        else: # Matriz continha apenas padding ou estava vazia de valores reais
            return ""

        reconstructed_text = "".join([chr(val) for val in values_no_padding])
        return reconstructed_text

# --- Exemplo Didático de Uso ---
if __name__ == "__main__":


    # Conversor padrão: minúsculas, normaliza espaços, remove pontuação
    conversor = SimpleTextMatrixConverter(width=20)

    print("Texto Original:")
    print(f"'{filecontent1}'\n")

    matriz_numerica, texto_limpo = conversor.text_to_matrix(filecontent1)

    print("Texto Limpo (usado para a matriz):")
    print(f"'{texto_limpo}'\n")

    print(f"Dimensões da Matriz: {matriz_numerica.shape}")
    print("Matriz Numérica:")
    print(matriz_numerica, "\n")

    texto_reconstruido = conversor.matrix_to_text(matriz_numerica)
    print("Texto Reconstruído da Matriz:")
    print(f"'{texto_reconstruido}'\n")

    print(f"Reconstrução bem-sucedida? {texto_limpo == texto_reconstruido}")
```
